```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 1. Write code to apply image blurring using Gaussian blur in OpenCV with Python and explain its effect on the image
img = cv2.imread(r"images\basket.jpg")
a = img.shape
print(a)

h, w, c = img.shape
print("Dimensions of the image is:\n\nHeight:", h, "pixels\nWidth:", w, "pixels\nNumber of Channels:", c)
print(img)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

fig, ax = plt.subplots(1, 2, figsize=(18, 18))
ax[0].imshow(gray, cmap='gray')
ax[1].imshow(blurred, cmap='gray')
ax[0].axis('off')
ax[1].axis('off')

# 2. How would you perform image inversion (negation) using OpenCV in Python, and what is its significance in image processing?
inverted = 255 - img

# Display the original and inverted images
cv2.imshow('Original Image', img)
cv2.imshow('Inverted Image', inverted)
cv2.waitKey(0)
cv2.destroyAllWindows()

# 3. Write a Python function using OpenCV to convert a colored image to a binary image
def colored_to_binary(threshold_value):
    # Convert the image to grayscale
    grayscale_image = cv2.cvtColor(colored_image, cv2.COLOR_BGR2GRAY)
    
    # Apply binary thresholding
    _, binary_image = cv2.threshold(grayscale_image, threshold_value, 255, cv2.THRESH_BINARY)
    
    return binary_image

colored_image = cv2.imread(r".\images\flowers.jpg")
threshold_value = 100
binary_image = colored_to_binary(threshold_value)

# Display the original and binary image
cv2.imshow('Original Image', colored_image)
cv2.imshow('Binary Image', binary_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# 4. How would you translate an image horizontally by a specified number of pixels using OpenCV in Python?
# Number of pixels to shift
shift_pixels = 100

# Get image dimensions
height, width = colored_image.shape[:2]

# Create an empty image with the same dimensions
translated_image = np.zeros_like(colored_image)

# Loop through each row of the image
for y in range(height):
    new_y = y
    if 0 <= y < height:
        new_y = (y + shift_pixels) % height
    translated_image[new_y, :] = colored_image[y, :]

# Display the original and translated images
cv2.imshow('Original Image', colored_image)
cv2.imshow('Translated Image', translated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
