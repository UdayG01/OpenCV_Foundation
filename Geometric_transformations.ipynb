```python
import cv2
import numpy as np

# Load the image
img = cv2.imread('.\images\lion.jpeg')

# Display the original image
cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Get the height and width of the image
h = img.shape[0]
w = img.shape[1]

# Define the translation matrix
tx = h / 4
ty = w / 4
translation_matrix = np.array([[1, 0, tx], [0, 1, ty]])

# Apply the translation to the image
translated_image = cv2.warpAffine(img, translation_matrix, (h, w))

# Display the translated image
cv2.imshow('Translated Image', translated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Get the height and width of the image
h, w = img.shape[:2]

# Define the rotation angle and center
angle = 45
center = (w / 2, h / 2)

# Define the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)

# Apply the rotation to the image
rotated_image = cv2.warpAffine(img, rotation_matrix, (w, h))

# Display the rotated image
cv2.imshow('Rotated Image', rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Define the scaling factor
scale = 0.5

# Apply scaling to the image
scaled_image = cv2.resize(img, (int(w * scale), int(h * scale)), interpolation=cv2.INTER_AREA)

# Display the scaled image
cv2.imshow('Scaled Image', scaled_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Define the shearing factor
shearing_factor = 0.5

# Define the shearing matrix
m = np.float32([[1, shearing_factor, 0], [0, 1, 0], [0, 0, 1]])

# Apply shearing to the image
sheared_image = cv2.warpPerspective(img, m, (img.shape[1], img.shape[0]))

# Display the sheared image
cv2.imshow('Sheared Image', sheared_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
