```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Python code to read image using OpenCV
img = cv2.imread(".\images\lion.jpeg")
a = img.shape
print(a)

h, w, c = img.shape
print("Dimensions of the image is:\n\nHeight:", h, "pixels\nWidth:", w, "pixels\nNumber of Channels:", c)
print(img)

cv2.imshow('Lion', img)
k = cv2.waitKey(0)
if k == 27 or k == ord('q'):
    cv2.destroyAllWindows()

# 2. Python code to show an input image using OpenCV in RGB colorspace
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.imwrite('.\output_images\Lion_RGB.jpeg', rgb)

# 3. Split the input image into R, G, and B channels
B, G, R = cv2.split(img)

# 4. Python code to show an input image using OpenCV in R channel and Green Channel
cv2.imshow('Red Channel', R)
k = cv2.waitKey(0)
if k == 27 or k == ord('q'):
    cv2.destroyAllWindows()

cv2.imshow('Green Channel', G)
k = cv2.waitKey(0)
if k == 27 or k == ord('q'):
    cv2.destroyAllWindows()

# 5. Python code to resize the input image
# Resize the image to half its original size
half = cv2.resize(img, (0, 0), fx=0.1, fy=0.1)

# Resize the image to a specific size (20, 20)
bigger20 = cv2.resize(img, (20, 20))

# Resize the image with linear interpolation
stretch_near = cv2.resize(img, (20, 20), interpolation=cv2.INTER_LINEAR)

# Display the resized images
Titles = ["Original", "Half", "Bigger", "Interpolation Nearest"]
images = [img, half, bigger20, stretch_near]

for i in range(len(images)):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])
plt.show()

print(bigger20)

# 6. Write a program in Python to demonstrate arithmetic operation on the input image
# a) Adding constant value (say 70) to each pixel of the input image in R channel and display the resultant image
Bn, Gn, Rn = cv2.split(bigger20)
print(Rn)

for i in range(20):
    for j in range(20):
        Rn[i][j] += 70

print(Rn)
plt.title("Red Channel Image with +70 pixel values")
plt.imshow(Rn)
plt.show()

# b) Subtract 100 from each pixel values in R channel and display the resultant image
for i in range(20):
    for j in range(20):
        Rn[i][j] -= 100

print(Rn)
plt.title("Red Channel Image with -100 pixel values")
plt.imshow(Rn)
plt.show()

# c) Take two different RGB images from two different paths and add two images using + operator. Show the final resultant image
img2 = cv2.imread("lion.jpeg")
rgb2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
smaller20 = cv2.resize(rgb2, (20, 20))

final = bigger20 + smaller20
plt.title("Image with + 2 images")
plt.imshow(final)
plt.show()

# 7. Convert RGB image into Grey Image and display the grey image. Now convert the grey image into binary image
gray = cv2.cvtColor(bigger20, cv2.COLOR_BGR2GRAY)
plt.title("Grey Image")
plt.imshow(gray)
plt.show()

print(gray)

for i in range(20):
    for j in range(20):
        if gray[i][j] >= 90:
            gray[i][j] = 100
        elif gray[i][j] < 90:
            gray[i][j] = 0

print(gray)
plt.title("Binary Image")
plt.imshow(gray)
plt.show()
```
